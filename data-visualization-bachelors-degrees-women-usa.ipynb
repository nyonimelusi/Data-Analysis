Let us generate line charts for four STEM degree and compare the rates of Men and Women graduating. We will use the matplotlib.pyplot to plot our data. What conclusion can you draw?

* Engineering and Computer Science have large gender gaps
* The Gap in Biology and Math & Statistics is quite small
* In Biology Women are getting to be a majority
* In Math & Statistics the rates are dropping after getting very close in 2000

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.plot(women_degrees["Year"], women_degrees["Biology"])
plt.show()
```

```python
plt.plot(women_degrees["Year"], women_degrees["Biology"], label = "Women")
plt.plot(women_degrees["Year"], 100-women_degrees["Biology"], label = "Men")
plt.legend(loc = "upper right")
plt.title("Degree awarded to Women")
plt.xlabel("Years")
plt.ylabel("Percentage")
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(women_degrees["Year"], women_degrees["Biology"], label = "Women", c = "red")
ax.plot(women_degrees["Year"], 100-women_degrees["Biology"], label = "Men", c = "green")
plt.legend(loc = "upper right")
ax.set_title("Degree awarded to Women")
plt.xlabel("Years")
ax.set_ylabel("Percentage")
ax.tick_params(left = False, bottom = False)
plt.show()
```

```python
maj_cats = ["Biology", "Computer Science", "Engineering", "Math and Statistics"]
```

```python
fig = plt.figure(figsize = (12,12))

for i in range(4):
    ax = fig.add_subplot(2,2,i+1)
    ax.plot(women_degrees["Year"], women_degrees[maj_cats[i]], label = "Women", c = "red",)
    ax.plot(women_degrees["Year"], 100-women_degrees[maj_cats[i]], label = "Men", c = "green")
    ax.tick_params(left = False, bottom = False)
    ax.set_title(maj_cats[i])
    ax.set_xlim(1968,2013)
    ax.set_ylim(0,100)
    
    for value in ax.spines.values():
        value.set_visible(False)
plt.legend()
plt.show()
```

I would like to compare the growth for every 5 years

```python
import seaborn as sns
```

```python
sns.barplot(x="Year",y="Agriculture",data = women_degrees)
```

```python
group = women_degrees.groupby("Year").mean().reset_index()
```

```python
def plotmatrix(start,end):
    fig, axs = plt.subplots(nrows = 2, ncols=2)
    i = 0
    cols = women_degrees.columns[start:end]
    fig.set_size_inches(14, 10)
    for indi in range(2):
        for indj in range(2):
                sns.barplot(x="Year",y=str(cols[i]),data = group,ax = axs[indi][indj],\
                            order = ['71-76', '76-81', '81-86', '86-91', '91-96','96-01','01-06','06-11'])
                i+=1
                #plt.xticks(rotation = 90)
```

```python
plotmatrix(1,5)
```

So from the above fig, it's clear that

1. Growth is in increasing order in Agriculture Field and even in Architecture Field
1. In Art and Performance, the growth is almost stagnant.
1. It's interesting to see that growth for Biology is declining from 2001-06 to 2006-11.

```python
plotmatrix(6,10)
```

It's clearly understood that, women growth in Compute Science is declining

```python
plotmatrix(11,15)
```
